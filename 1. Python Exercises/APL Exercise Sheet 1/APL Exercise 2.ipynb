```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

(a) Plotting the cosine function

```python
time = np.linspace(0,2,500)
plt.figure(figsize=(12,3))
plt.title("Cosine Plot")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (units)")
plt.plot(time, np.cos(2*np.pi*10*time+np.pi/6));
```

(b) Plotting two cosine functions on the same plot

```python
time = np.linspace(0,2,500)
plt.figure(figsize=(12,3))
plt.title("Two Cosine Plot")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (units)")
plt.plot(time, np.cos(2*np.pi*10*time+np.pi/6), color='b')
plt.plot(time, np.cos(2*np.pi*8*time+np.pi/3), color='r');
```

(c) Determining and plotting resulting wave when (b) waves interfere

```python
time = np.linspace(0,2,500)
plt.figure(figsize=(12,3))
plt.title("Combined Cosine Plot")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (units)")
combined_wave = np.cos(2*np.pi*10*time+np.pi/6)+0.8*np.cos(2*np.pi*8*time+np.pi/3)
plt.plot(time, combined_wave, color='purple');
```

(d) Frequency spectrum of FFT investigation

```python
wave_fft = np.fft.fft(combined_wave)

half_wave_fft = np.abs(wave_fft[0:int(len(wave_fft)/2)])
# print(half_wave_fft)

plt.subplot(211)
plt.title("Fast Fourier Transform Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (Units)")
plt.plot(half_wave_fft)
plt.subplot(212)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amfaplitude (Units)")
plt.xlim(10,25)
plt.plot(half_wave_fft);
```

We expect 2 peaks because the input function is the combination of 2 cosine functions. The differing amplitudes in the fft plot corresponds to the differing amplitudes of the original cosine functions.
We have a time domain of 2 so we expect our output frequencies to be twice their original value (as observed).
Our domain is halved (250 when compared to 500), due to the nature of the fft (its symmetricality).
